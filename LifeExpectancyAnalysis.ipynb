## Question: Which countries have the highest and lowest Life expectancy?
- Pull relevant information -> Year, Country, Gender, Life Expectancy
- Determine axis and which type of graph(s) to use
- will use oldest and most recent year for year column and filter gender column by 'Both sexes'

```python
oldest_year = df["Year"].min()
print(oldest_year)
```

```python
latest_year = df["Year"].max()
print(latest_year)
```

Create two separate tables for countries that have the highest/lowest life expectancy in 1990 and 2019 for both genders

```python
df_1990_both_sexes = df[df["Year"].isin([1990]) & df["Gender"].isin(["Both sexes"])]
df_2019_both_sexes = df[df["Year"].isin([2019]) & df["Gender"].isin(["Both sexes"])]
```

```python
highest_life_1990 = df_1990_both_sexes.nlargest(10, "Life Expectancy")
lowest_life_1990 = df_1990_both_sexes.nsmallest(10, "Life Expectancy")
highest_life_2019 = df_2019_both_sexes.nlargest(10, "Life Expectancy")
lowest_life_2019 = df_2019_both_sexes.nsmallest(10, "Life Expectancy")
```

**Graph of the countries with highest life expectancy in 1990**

```python
highest_life_1990.plot.scatter(x = "Country", y = "Life Expectancy" , title = "Country vs. Life Expectancy in Years(1990)")
plt.xticks(rotation = 90);
```

**Graph of the countries with the lowest life expectancy in 1990**

```python
lowest_life_1990.plot.scatter(x = "Country", y = "Life Expectancy", title = "Country vs. Life Expectancy in Years(1990)")
plt.xticks(rotation = 90);
```

**Graph of countries with highest life expectancy in 2019**

```python
highest_life_2019.plot.scatter(x = "Country", y = "Life Expectancy", title = "Country vs. Life Expectancy in Years(2019)")
plt.xticks(rotation = 90);
```

**Graph of countries with lowest life expectancy in 2019**

```python
lowest_life_2019.plot.scatter(x = "Country", y = "Life Expectancy", title = "Country vs. Life Expectancy in Years(2019)")
plt.xticks(rotation = 90);
```

**Insights:** Based on the graphs, It appears that lesser developed countries have a lower life expectancy, while more developed countries have a higher life expectancy. This makes sense because there's more resources available for people in first world countries.

## Question: Does diet affect life expectancy?
- Pull relevant information -> Country, Diet Calories Plant Protein, Animal Protein, Diet Calories Fat, Diet Calories Carbohydrates
- Determine axis and which type of graph(s) to use
- Diet metrics will be centered around macronutrients
- Use 'Both sexes' for gender metric

```python
df_both_sexes = df[df["Gender"].isin(["Both sexes"])]
```

**Graph of countries life expectancy vs their carbohydrate intake(kcal)**

```python
df_both_sexes.plot.scatter(x = "Diet Calories Carbohydrates" , y = "Life Expectancy", title = "Carbohydrate intake(kcal) vs. Life Expectancy in years",figsize=(10,5), alpha = .4);
```

**Insights**: There is slight correlation between carb intake and life expectancy. Many of the countries with higher life expectancies tend to aggregate towards an intake of 1750 kcal of carbs. However, lower carb intake doesnt appear to negatively affect life expectancy.

**Graph of countries life expectancy vs fat intake(kcal)**

```python
df_both_sexes.plot.scatter(x = "Diet Calories Fat", y = "Life Expectancy" , title = "Fat intake(kcal) vs. Life Expectancy in years", figsize = (10,6), alpha = .6);
```

**Insights:** There is a clear correlation between between the amount of fat calories consumed and life expectancy. Countries with higher fat intake tend to have a higher life expectancy.

**Graph of countries life expectancy vs plant protein intake(kcal)**

```python
df_both_sexes.plot.scatter(x = "Diet Calories Plant Protein", y = "Life Expectancy", title = "Plant Protein Intake vs. Life Expectancy in Years", figsize = (10,6), alpha = .4); 
```

**Insights:** There seems to be no correlation between plant protein intake and life expectancy. This could be due to a number of reasons such that plant protein isnt as nutrient dense as protein sources from meat.

**Graph of countries life expectancy vs animal protein intake(kcal)**

```python
df_both_sexes.plot.scatter(x = "Diet Calories Animal Protein", y = "Life Expectancy", title = "Animal Protein Intake(kcal) vs. Life Expectancy", figsize = (10,5), alpha = .4);
```

**Insights:** Animal Protein Intake and life expectancy are strongly correlated. Countries that consume more animal protein have a higher life expectancy.

## Question: Is there a correlation between alcohol abuse and life expectancy?
- pull relevant information --> Alcohol Abuse, Life Expectancy, Gender, Year
- determine which type of graph(s) to use
- determine what alcohol abuse represents

Determining what alcohol abuse represents as a value

```python
df["Alcohol Abuse"].dtype
```

The data type of the alcohol abuse is a float, so it represents a percentage of the population

removing null values and filtering 'Gender' by 'Both sexes'

```python
alcohol_df = df.dropna(subset = ["Alcohol Abuse"])
```

```python
alcohol_df = alcohol_df.loc[(alcohol_df["Gender"] == "Both sexes")]
```

Graph of Alcohol Abuse vs. Life Expectancy for both genders in 2019

```python
alcohol_df.plot.scatter(x = "Alcohol Abuse" , y = "Life Expectancy", title = "Alcohol Abuse(%) vs. Life Expectancy in years", fig = (10,5), alpha = .4);
```

**Insights:** This is surprising, but there isn't a strong correlation between alcohol abuse and life expectancy. This could be caused by the notion that countries with higher alcohol abuse have better healtchare/living conditions.
